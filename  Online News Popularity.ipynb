<h1 style="color:purple;">Predicting the number of times a news article is shared</h1>

&nbsp;

<h3 style="color:maroon;"><em>Problem statement :</em></h3>

This dataset summarizes a heterogeneous set of features about articles published by Mashable in a period of two years. The goal is to predict the number of shares in social networks (popularity).

<h3 style="color:darkblue;"><em>4. Building a network to predict shares</em></h3>

<h4 style="color:darkblue;"><em>4.1 Splitting data into traind test</em></h4>

```python
x = data.drop([' shares','url'],axis=1).values
y = data[' shares'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
x_train =np.asarray(x_train).astype(np.float32)
x_test =np.asarray(x_test).astype(np.float32)
y_train =np.asarray(y_train).astype(np.float32)
y_test =np.asarray(y_test).astype(np.float32)
```

<h4 style="color:darkblue;"><em>4.2 Creating a sequential model architechture </em></h4>

```python
epochs = 50
```

```python
tf.keras.backend.clear_session()
```

```python
model = Sequential()

model.add(Dense(64,activation='relu',input_dim=len(x_train[0])))
model.add(Dense(128,activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(1))

model.summary()
```

```python
optimizer = tf.keras.optimizers.Adam()

model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
```

<h4 style="color:darkblue;"><em>4.3 Training and results </em></h4>

```python
history = model.fit(x_train,y_train,epochs=epochs, validation_data=(x_test,y_test),verbose=1)
```

Reference : https://machinelearningmastery.com/display-deep-learning-model-training-history-in-keras/

```python
# summarize history for accuracy
plt.plot(history.history['mse'])
plt.plot(history.history['val_mse'])
plt.title('model MSE')
plt.ylabel('MSE')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
